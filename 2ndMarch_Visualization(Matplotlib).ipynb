```python
import numpy as np
import matplotlib.pyplot as plot
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
```

```python
x
```

```python
y
```

```python
plot.scatter(x, y)
```

```python
plot.figure(figsize = (6,2))
plot.scatter(x, y, c = 'r', marker = 'P')
plot.xlabel('This is my x axis')
plot.ylabel('This is my y axis')
plot.title('Scatter Plot')
plot.grid()
```

```python
plot.bar(x,y)
```

```python
x = ['a', 'b','c', 'd', 'e']
```

```python
y = np.random.rand(5)
```

```python
plot.title('this is my first ever bar graph')
plot.bar(x,y)
plot.xlabel('This is my x axis')
plot.ylabel('This is my y axis')
plot.grid()
```

```python
plot.title('this is my first ever bar graph')
plot.barh(x, y)
plot.xlabel('This is my x axis')
plot.ylabel('This is my y axis')
plot.grid()
```
