<h1>Standardization</h1>

```python
mean = np.mean(df['total_bill'])
std = np.std(df['total_bill'])
standardized_data = []
for i in list(df['total_bill']):
    zscore = i - mean / std
    standardized_data.append(zscore)
```

<h1>Normalization</h1>
<h2>Min Max Scaler</h2>

```python
df = sns.load_dataset('taxis')
```

```python
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
normalize = MinMaxScaler()
```

```python
normalize.fit(df[['distance']])
```

```python
df['normalised_distance'] = pd.DataFrame(normalize.transform(df[['distance']]))
```

```python
df.head()[['distance','normalised_distance','fare','tip']]
```

```python
df['normalised_total']=pd.DataFrame(normalize.fit_transform(df[['total']]))
```
