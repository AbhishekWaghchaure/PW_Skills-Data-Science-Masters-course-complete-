Data Interpolation techniques

Linear interpolation
polynomial interpolation
cubic interpolation

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])
```

```python
sns.scatterplot(x = x, y= y , c='red')
```

Linear Interpolation

```python
x_new = np.linspace(1,5,10) ##created new x values
y_interp = np.interp(x_new,x,y)## created y interpolated points
```

```python
sns.scatterplot(x = x_new, y= y_interp , c='red')
```

Cubic interpolation

```python
import pandas as pd

# create a sample dataset with missing values
data = {'time': ['12:00', '13:00', '14:00', '15:00', '16:00'],
        'temperature': [20, 21, None, 23, 24]}
df = pd.DataFrame(data)

# # fill missing values using linear interpolation
# df['temperature'] = df['temperature'].interpolate()

# # print the updated dataset
# print(df)
```

```python
sns.scatterplot(x= df['time'], y = df['temperature'], c= 'red')
```

```python

# fill missing values using linear interpolation
df['temperature'] = df['temperature'].interpolate(method= 'cubic')
```

```python
df
```

```python
sns.scatterplot(x= df['time'], y = df['temperature'], c= 'red')
```

Polynomial data interpolation

```python
x = np.array([1,2,3,4,5])
y = np.array([2,4,9,16,25])
```

```python
x_new = np.linspace(1,5,10)
```

```python
f = np.polyfit(x,y,3)
```

```python
y_interp = np.polyval(f,x_new)
```

```python
sns.scatterplot(x=x_new,y=y_interp)
```
