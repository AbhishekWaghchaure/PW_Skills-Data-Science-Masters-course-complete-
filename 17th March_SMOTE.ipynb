```python
from sklearn.datasets import make_classification
import pandas as pd
```

```python
X,y = make_classification(n_samples=1000, n_redundant=0, n_features=2,n_clusters_per_class=1,weights=[0.90],random_state=12)
```

```python
df1 = pd.DataFrame(X,columns=['feature 1','feature 2'])
df2 = pd.DataFrame(y, columns=['target'])
final_df = pd.concat([df1,df2], axis = 1)
final_df
```

```python
import matplotlib.pyplot as plt
plt.scatter(final_df['feature 1'], final_df['feature 2'], c = final_df['target'])
```

```python
X,y = oversample.fit_resample(final_df[['feature 1','feature2']], final_df['target'])
```
