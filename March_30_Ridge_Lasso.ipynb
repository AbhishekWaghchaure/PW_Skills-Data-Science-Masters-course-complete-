Handling null values

```python
dataset.isnull().sum()
```

```python
dataset[dataset.isnull().any(axis = 1)]
```

```python
dataset.loc[:122,"Region"]=0
dataset.loc[122:,"Region"]=1
```

```python
dataset.info()
```

```python
dataset.head()
```

```python
dataset['Region'] = dataset['Region'].astype(int)
```

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
dataset = dataset.dropna().reset_index(drop=True)
```

```python
dataset.isnull().sum()
```

```python
dataset.iloc[[122]]
```

```python
dataset = dataset.drop(122).reset_index(drop = True)
```

```python
dataset.iloc[122]
```

```python
dataset.columns
```

```python
dataset.columns = dataset.columns.str.strip()
```

```python
dataset.columns
```

```python
dataset.info()
```

```python
dataset.head()
```

```python
dataset[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']] = dataset[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
dataset.info()
```

```python
objects = [i for i in dataset.columns if dataset[i].dtype == 'O' ]
```

```python
objects
```

```python
for i in objects:
    if i != 'Classes':
        dataset[i] = dataset[i].astype(float)
```

```python
dataset.info()
```

```python
dataset.to_csv('Algerian_forest_fires_cleaned_data.csv', index = False)
```
